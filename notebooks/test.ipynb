```python
import pandas as pd

tas = pd.read_csv("../data/state_annual_tas_cmip6_2020_2070.csv")
heat = pd.read_csv("../data/state_annual_extreme_heat_cmip6_2020_2070.csv")

print("TAS columns:")
print(tas.columns)

print("\nHeat columns:")
print(heat.columns)

print("\nTop average warming, 2070, ssp585:")
print(
    tas[(tas["year"] == 2070) & (tas["scenario"] == "ssp585")]
    .sort_values("tas_change_from_2020_c", ascending=False)
    [["state", "tas_change_from_2020_c"]]
    .head(10)
)

print("\nTop additional 35C+ days, 2070, ssp585:")
print(
    heat[(heat["year"] == 2070) & (heat["scenario"] == "ssp585")]
    .sort_values("hot_days_35c_change_from_2020", ascending=False)
    [["state", "hot_days_35c", "hot_days_35c_change_from_2020"]]
    .head(10)
)
```
